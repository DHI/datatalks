```python
df_K14 = mikeio.read(r"C:\Users\jem\Source\datatalks\talk_9_reusable_code\data\obs\K14.dfs0").to_dataframe()
```

```python
from tsod import RangeDetector, GradientDetector, CombinedDetector
dd = CombinedDetector([RangeDetector(0,10), GradientDetector(0.00083)])
```

```python
anomalies = dd.detect(df_K14.Hm0)
plt.plot(df_K14)
plt.plot(df_K14[anomalies], 'ro', label='Anomaly');
plt.show()
df_K14[anomalies.values] = np.nan
```
