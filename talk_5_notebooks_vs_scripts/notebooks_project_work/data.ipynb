# Loading and preparing data

get raw data from ASCII file 

```python
fn = 'raw_Blankenese_WL.txt'
df = pd.read_csv(fn, sep='\s+', header=0, names=['Date', 'Time', 'WL [cm]'], usecols= [0,1,2], 
    parse_dates=[['Date', 'Time']], index_col='Date_Time')
```

```python
df.plot();
```

select daterange and make sure it comprises only equidistant timesteps of 5 mins

```python
df_cut = df['2016':'2017']
df_cut = df_cut.resample('5T').mean()
df_cut.plot();
```

convert from cm to m

```python
df_cut = df_cut/100
df_cut.plot();
```
